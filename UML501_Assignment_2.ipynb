```python
# =========================================
# UML501 - ASSIGNMENT 2 (FINAL COMPLETE)
# =========================================

import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics.pairwise import cosine_similarity

# =========================================
# LOAD DATA
# =========================================

df = pd.read_csv("Bike_Buyers.csv")  # upload this in Colab

print("Original Shape:", df.shape)

# =========================================
# PART I: FEATURE SELECTION + DATA TYPES
# =========================================

# Selected important features
features = [
    "Age", "Yearly Income", "Gender", "Marital Status",
    "Number of Children", "Education", "Occupation",
    "Commute Distance", "Region", "Purchased Bike"
]

df = df[features]

print("\nSelected Features:")
print(df.head())

# Data Types Classification (for report)
data_types = {
    "Age": "Continuous (Ratio)",
    "Yearly Income": "Continuous (Ratio)",
    "Gender": "Nominal",
    "Marital Status": "Nominal",
    "Number of Children": "Discrete",
    "Education": "Ordinal",
    "Occupation": "Nominal",
    "Commute Distance": "Ordinal",
    "Region": "Nominal",
    "Purchased Bike": "Binary"
}

print("\nData Types:")
for k, v in data_types.items():
    print(f"{k}: {v}")

# =========================================
# PART II: PREPROCESSING
# =========================================

# 1. Handle Missing Values
df.fillna(df.mode().iloc[0], inplace=True)

# 2. Discretization (Binning)
df["Income_Bin"] = pd.cut(df["Yearly Income"], bins=5, labels=False)

# 3. Normalization
scaler_minmax = MinMaxScaler()
df[["Age_norm", "Income_norm"]] = scaler_minmax.fit_transform(df[["Age", "Yearly Income"]])

# 4. Standardization
scaler_std = StandardScaler()
df[["Age_std", "Income_std"]] = scaler_std.fit_transform(df[["Age", "Yearly Income"]])

# 5. One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=[
    "Gender", "Marital Status", "Occupation", "Region"
], drop_first=True)

print("\nAfter Encoding Shape:", df_encoded.shape)

# =========================================
# PART III: SIMILARITY MEASURES
# =========================================

print("\n===== SIMILARITY MEASURES =====")

# Take first two rows
row1 = df_encoded.iloc[0].values
row2 = df_encoded.iloc[1].values

# 1. Simple Matching Coefficient
smc = np.sum(row1 == row2) / len(row1)
print("Simple Matching:", smc)

# 2. Jaccard Similarity (binary only)
binary1 = (row1 > 0).astype(int)
binary2 = (row2 > 0).astype(int)

intersection = np.sum((binary1 == 1) & (binary2 == 1))
union = np.sum((binary1 == 1) | (binary2 == 1))

jaccard = intersection / union if union != 0 else 0
print("Jaccard Similarity:", jaccard)

# 3. Cosine Similarity
cos_sim = cosine_similarity([row1], [row2])[0][0]
print("Cosine Similarity:", cos_sim)

# =========================================
# CORRELATION
# =========================================

print("\n===== CORRELATION =====")

corr = df["Commute Distance"].astype("category").cat.codes.corr(df["Yearly Income"])
print("Correlation (Commute Distance vs Income):", corr)

# =========================================
# END
# =========================================
```
