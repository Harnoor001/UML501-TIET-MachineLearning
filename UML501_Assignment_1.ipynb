```python
# =========================================
# UML501 - ASSIGNMENT 1 (FINAL)
# =========================================

import numpy as np
from PIL import Image

# =========================================
# Q1: BASIC NUMPY
# =========================================

print("\n===== Q1 =====")

# (a) Reverse array
arr = np.array([1, 2, 3, 6, 4, 5])
print("Reversed:", arr[::-1])

# (b) Flatten using two methods
array1 = np.array([[1,2,3],[2,4,5],[1,2,3]])
print("Flatten (method 1):", array1.flatten())
print("Flatten (method 2):", array1.ravel())

# (c) Compare arrays
arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[1,2],[3,4]])
print("Equal:", np.array_equal(arr1, arr2))

# (d) Most frequent value + indices
x = np.array([1,2,3,4,5,1,2,1,1,1])
values, counts = np.unique(x, return_counts=True)
max_val = values[np.argmax(counts)]
indices = np.where(x == max_val)
print("Most frequent:", max_val)
print("Indices:", indices)

y = np.array([1,1,1,2,3,4,2,4,3,3])
values, counts = np.unique(y, return_counts=True)
max_val = values[np.argmax(counts)]
indices = np.where(y == max_val)
print("Most frequent (y):", max_val)
print("Indices:", indices)

# (e) Matrix operations
gfg = np.matrix('[4,1,9;12,3,1;4,5,6]')
print("Total sum:", np.sum(gfg))
print("Row-wise sum:", np.sum(gfg, axis=1))
print("Column-wise sum:", np.sum(gfg, axis=0))

# (f) Matrix calculations
n_array = np.array([[55,25,15],[30,44,2],[11,45,77]])
print("Diagonal sum:", np.trace(n_array))
print("Eigenvalues:", np.linalg.eigvals(n_array))
print("Eigenvectors:", np.linalg.eig(n_array)[1])
print("Inverse:\n", np.linalg.inv(n_array))
print("Determinant:", np.linalg.det(n_array))

# (g) Matrix multiplication + covariance
p = np.array([[1,2],[2,3]])
q = np.array([[4,5],[6,7]])
print("Multiplication:\n", np.dot(p,q))
print("Covariance:\n", np.cov(p, q))

p = np.array([[1,2],[2,3],[4,5]])
q = np.array([[4,5,1],[6,7,2]])
print("Multiplication:\n", np.dot(p,q))

# (h) Inner, Outer, Cartesian
x = np.array([[2,3,4],[3,2,9]])
y = np.array([[1,5,0],[5,10,3]])

print("Inner:", np.inner(x,y))
print("Outer:", np.outer(x,y))

# Cartesian product
cartesian = np.array(np.meshgrid(x.flatten(), y.flatten())).T.reshape(-1,2)
print("Cartesian:\n", cartesian)

# =========================================
# Q2: NUMPY MATH & STATS
# =========================================

print("\n===== Q2 =====")

array = np.array([[1,-2,3],[-4,5,-6]])

# (i) Absolute
print("Absolute:\n", np.abs(array))

# (ii) Percentiles
flat = array.flatten()
print("Percentiles (flat):", np.percentile(flat, [25,50,75]))
print("Percentiles (column):", np.percentile(array, [25,50,75], axis=0))
print("Percentiles (row):", np.percentile(array, [25,50,75], axis=1))

# (iii) Mean, Median, Std
print("Mean flat:", np.mean(flat))
print("Mean col:", np.mean(array, axis=0))
print("Mean row:", np.mean(array, axis=1))

print("Median flat:", np.median(flat))
print("Std flat:", np.std(flat))

# (b) Floor, ceil, trunc, round
a = np.array([-1.8,-1.6,-0.5,0.5,1.6,1.8,3.0])
print("Floor:", np.floor(a))
print("Ceil:", np.ceil(a))
print("Trunc:", np.trunc(a))
print("Round:", np.round(a))

# =========================================
# Q3: SEARCHING & SORTING
# =========================================

print("\n===== Q3 =====")

array = np.array([10,52,62,16,16,54,453])

print("Sorted:", np.sort(array))
print("Indices:", np.argsort(array))
print("4 smallest:", np.sort(array)[:4])
print("5 largest:", np.sort(array)[-5:])

array = np.array([1.0,1.2,2.2,2.0,3.0,2.0])

print("Integer elements:", array[array.astype(int) == array])
print("Float elements:", array[array.astype(int) != array])

# =========================================
# Q4: IMAGE TO ARRAY FUNCTION
# =========================================

print("\n===== Q4 =====")

def img_to_array(path):
    img = Image.open(path)
    arr = np.array(img)

    if len(arr.shape) == 3:
        print("RGB Image")
    else:
        print("Grayscale Image")

    # Save as text
    np.savetxt("image_array.txt", arr.reshape(-1, arr.shape[-1] if len(arr.shape)==3 else 1))
    print("Saved as image_array.txt")

    return arr

# Example usage (upload image in Colab)
# img_array = img_to_array("test.jpg")

# Load saved file
try:
    loaded = np.loadtxt("image_array.txt")
    print("Loaded array shape:", loaded.shape)
except:
    print("Run img_to_array first to generate file")

# =========================================
# END
# =========================================
```
